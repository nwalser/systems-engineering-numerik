```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure


# 1. load data
data = np.genfromtxt('SlitCuring.100Imprints.15mm.txt')

t = data[:,0]
s1 = data[:,8]
s2 = data[:,18]

# visualize data
figure(figsize=(30, 6), dpi=80)
plt.plot(t, s1)
plt.plot(t, s2)
plt.title("Slit Positions")
plt.xlabel("Tick")
plt.ylabel("Position [mm]")
plt.show()
```

```python
figure(figsize=(30, 6), dpi=80)
plt.plot(t, np.absolute(s1-s2))
plt.title("Slit Width")
plt.xlabel("Tick")
plt.ylabel("Width [mm]")
plt.show()
```
