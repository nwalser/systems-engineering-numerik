```python
import numpy as np
import matplotlib.pyplot as plt

def fibonacci():
    previous=0
    current=1

    yield previous

    while(True):
        yield current
        previous, current = current, previous + current

generator = fibonacci()
fib = np.array([next(generator) for idx in range(1000)])

plt.plot(fib)
plt.show()

# datatype is always object because values are too large to be stored as np.int
print('Datatype fib:', fib.dtype)
print('Datatype fib*0.5:', (fib*0.5).dtype)
print('Datatype fib*5:', (fib*5).dtype)
print('Datatype fib*5.0:', (fib*5.0).dtype)
```
